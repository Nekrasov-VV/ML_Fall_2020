```python
%matplotlib inline
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train-2.csv')
```

```python
data.info()
```

```python
data.loc[(data['TARGET'] == 'low'), 'TARGET'] = 0
data.loc[(data['TARGET'] == 'medium'), 'TARGET'] = 1
data.loc[(data['TARGET'] == 'high'), 'TARGET'] = 2
data.head()
```

```python
def code_mean(data, cat_feature, real_feature):
    return (data[cat_feature].map(data.groupby(cat_feature)[real_feature].sum()))
def count_elements(str1):
    k = 0
    for s in str1:
        if s == '\'':
            k+=1
    return k//2;
```

```python
data.fillna(value='', inplace=True)
data.drop(columns=[ 'building_id', 'Id', 'listing_id','street_address', 'created', 'manager_id'], inplace = True)
```

```python
for i in range(len(data['description'])):
    data.loc[i,'description'] = len(data.loc[i,'description'])
    if(len(data.loc[i,'display_address']) > 0):
        data.loc[i,'display_address'] = 1
    else:
        data.loc[i,'display_address'] = 0
    data.loc[i,'photos'] = count_elements(data.loc[i,'photos'])
    data.loc[i,'features'] = count_elements(data.loc[i,'features'])
```

```python
X = data.drop(columns=['TARGET']) 
y = data['TARGET']
```

```python
for col in X.columns:
    X[col] = (X[col] - X[col].mean()) / X[col].std()
X.head()
```

```python
def date_parser(datestr):
    return pd.datetime.strptime(datestr, '%Y-%m-%d %H:%M:%S')
data['created'].apply(date_parser)
data['created'] = pd.to_datetime(data['created'], infer_datetime_format=True)
```
